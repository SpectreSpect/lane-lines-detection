```python
import yaml
import os
import re
import pandas as pd
```

```python
with open("../data/russian-traffic-signs-recognition/data.yaml", "r") as file:
    config = yaml.safe_load(file)
```

```python
label_names = config["names"]
```

```python
# labels_dir = "../data/russian-traffic-signs-recognition/train/labels"
```

```python
# label_file_names = os.listdir(labels_dir)
```

```python
def get_annotations_df(labels_dir: str):
    labels = []
    boxes = []
    image_names = []
    label_file_names = os.listdir(labels_dir)
    for label_file_name in label_file_names:
        label_file_path = os.path.join(labels_dir, label_file_name)
        with open(label_file_path, 'r') as file:
            for line in file.readlines():
                values = re.findall(r'\d+\.\d+|\d+', line)

                label = int(values[0])
                x = float(values[1])
                y = float(values[2])
                width = float(values[3])
                height = float(values[4])

                xtl = x - width / 2
                ytl = y - height / 2
                xbr = x + width / 2
                ybr = y + width / 2

                labels.append(label)
                boxes.append([xtl, ytl, xbr, ybr])
                image_names.append(label_file_name)  

    df = pd.DataFrame({"label": labels, "box": boxes, "image_name": image_names})
    return df
```

```python
df_train = get_annotations_df("../data/russian-traffic-signs-recognition/train/labels")
df_valid = get_annotations_df("../data/russian-traffic-signs-recognition/valid/labels")
df_test = get_annotations_df("../data/russian-traffic-signs-recognition/test/labels")
df_train["label_name"] = df_train["label"].map(lambda value: label_names[value])
df_valid["label_name"] = df_train["label"].map(lambda value: label_names[value])
df_test["label_name"] = df_train["label"].map(lambda value: label_names[value])

df = pd.concat([df_train, df_valid, df_test], ignore_index=True)

# pd.set_option('display.max_rows', None)
# df_train["label_name"].value_counts()
```
